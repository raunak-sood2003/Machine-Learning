```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import metrics
```

```python
#One Hot Encoded Model
```

```python
train = 'C:\\Users\\rrsoo\\Advanced Stuff\\Titanic\\train.csv'
test = "C:\\Users\\rrsoo\\Advanced Stuff\\Titanic\\test.csv"
gender_submission = "C:\\Users\\rrsoo\\Advanced Stuff\\Titanic\\gender_submission.csv"
```

```python
read_training_data = pd.read_csv(train)
y_train = read_training_data.Survived
read_training_data = read_training_data.drop(['Name', 'PassengerId', 'Survived', 'Ticket', 'Cabin'], axis = 'columns') #Name and PassengerId irrelavent to survival
read_training_data = read_training_data
read_testing_data = pd.read_csv(test)
read_testing_data = read_testing_data.drop(['Name', 'PassengerId', 'Ticket', 'Cabin'], axis = 'columns')
read_submission_data = pd.read_csv(gender_submission)
read_training_data
```

```python
read_testing_data
```

```python
#Setting up dummy variables for training data
dummies2_train = pd.get_dummies(read_training_data.Sex)
#dummies3_train = pd.get_dummies(read_training_data.Ticket)
#dummies4_train = pd.get_dummies(read_training_data.Cabin)
dummies5_train = pd.get_dummies(read_training_data.Embarked)
#Setting up dummy variables for testing data
dummies2_test = pd.get_dummies(read_testing_data.Sex)
#dummies3_test = pd.get_dummies(read_testing_data.Ticket)
#dummies4_test = pd.get_dummies(read_testing_data.Cabin)
dummies5_test = pd.get_dummies(read_testing_data.Embarked)
dummies2_train
```

```python
#Merging the training data
merged_train = pd.concat([read_training_data, dummies2_train, 
                          dummies5_train], axis = 'columns') #Removed Dummies 3,4 (Ticket)
#Merging the testing data
merged_test = pd.concat([read_testing_data, dummies2_test,
                          dummies5_test], axis = 'columns')
merged_train
```

```python
#Final csv file for training data
final_train = merged_train.drop(['Sex', 'Embarked'], axis = 'columns')
final_train
```

```python
#Final csv file for testing data
final_test = merged_test.drop(['Sex', 'Embarked'], axis = 'columns')
final_test
```

```python
#Training data values
features = final_train.columns
X_train = final_train[features]
X_train = X_train.fillna(0)
#Testing data values
features2 = final_test.columns
X_test = final_test[features2]
X_test = X_test.fillna(0)
```

```python
model = DecisionTreeClassifier(random_state = 4)
model.fit(X_train,y_train)
```

```python
y_preds = model.predict(X_test)
y_preds
```
