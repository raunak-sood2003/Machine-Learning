```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image_dir = 'C:/Users/rrsoo/AdvancedStuff/Udacity_CV/capitol.jpg'
img = cv2.imread(image_dir)
img_copy = np.copy(img) #Always make copies
img_copy = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_copy)
```

```python
#High pass filters

gray = cv2.cvtColor(img_copy, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap = 'gray')
```

```python
sobel_x = np.array([[-1, 0, 1],
           [-2, 0, 2],
           [-1, 0, 1]])

filtered_image = cv2.filter2D(gray, -1, sobel_x) #vertical edge detection
plt.imshow(filtered_image)
```

```python
sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])
filtered_image2 = cv2.filter2D(gray, -1, sobel_y)
plt.imshow(filtered_image2)
```

```python
#What if you want the most prominent edges?...thresholding
threshold = 100
make_pixel_val = 255
f, thresh_img = cv2.threshold(filtered_image, threshold, make_pixel_val, cv2.THRESH_BINARY)
plt.imshow(thresh_img)
```

```python
#Low pass filters
image_dir = 'C:/Users/rrsoo/AdvancedStuff/Udacity_CV/brain_noise.jpg'
brain = cv2.imread(image_dir)
brain_copy = np.copy(brain)
brain_copy = cv2.cvtColor(brain, cv2.COLOR_BGR2RGB)
plt.imshow(brain_copy)
```

```python
#Gaussian blurring
filter_size = (5,5)
gauss_brain = cv2.GaussianBlur(brain_copy, filter_size, 0) #Last value is std, which does it on its own when 0
f, (axis1, axis2) = plt.subplots(1, 2, figsize = (16, 8))
axis1.set_title("Original Brain")
axis1.imshow(brain)

axis2.set_title("Brain with Low Pass")
axis2.imshow(gauss_brain)
```

```python
#Let's see how they compare with sobel filters
sobel_x = np.array([[-1,0,1],
           [-2,0,2],
           [-1,0,1]])

sobel_orig_brain = cv2.filter2D(brain, -1, sobel_x)
sobel_gauss_brain = cv2.filter2D(gauss_brain, -1, sobel_x)

f, (axis1, axis2) = plt.subplots(1, 2, figsize = (16, 8))
axis1.set_title("Original Brain + High Pass")
axis1.imshow(sobel_orig_brain)

axis2.set_title("Brain with Gaussian + High Pass")
axis2.imshow(sobel_gauss_brain)
```

```python
#Canny edge detection: Gaussian Blur --> Sobel Filter --> Non-max suppression --> Hysterisis thresholding

sunflower_dir = 'C:/Users/rrsoo/AdvancedStuff/Udacity_CV/sunflower.jpg'
sunflower = cv2.imread(sunflower_dir)
sunflower_copy = np.copy(sunflower)
sunflower_copy = cv2.cvtColor(sunflower, cv2.COLOR_BGR2GRAY)
plt.imshow(sunflower_copy, cmap = 'gray')
```

```python
lower_thresh = 120
upper_thresh = 240
canny_sunflower = cv2.Canny(sunflower_copy, lower_thresh, upper_thresh)
plt.imshow(canny_sunflower)
```

```python
phone_dir = 'C:/Users/rrsoo/AdvancedStuff/Udacity_CV/phone.jpg'
phone = cv2.imread(phone_dir)
phone_copy = np.copy(phone)
phone_copy = cv2.cvtColor(phone, cv2.COLOR_BGR2GRAY)
plt.imshow(phone_copy, cmap = 'gray')
```

```python
lower = 120
higher = 240
canny_phone = cv2.Canny(phone_copy, lower, higher)
plt.imshow(canny_phone)
```

```python
#Hough Transforms to connect the edges
rho = 1
theta = np.pi/180
threshold = 60
max_line_length = 50
max_line_gap = 5

hough_line = cv2.HoughLinesP(canny_phone, rho, theta, threshold, np.array([]), 
                              max_line_length, max_line_gap)
line_copy = np.copy(hough_line)
#Now we have to put lines over our image
for line in hough_line:
    for x1, y1, x2, y2 in line:
        line_copy = cv2.line(line_copy, (x1, y1), (x2, y2), (255, 0, 0), 5)
plt.imshow(line_copy)
```
