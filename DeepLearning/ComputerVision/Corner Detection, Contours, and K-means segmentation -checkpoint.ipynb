```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
chess_dir = 'C:/Users/rrsoo/AdvancedStuff/Udacity_CV/chess.jpg'
chess = cv2.imread(chess_dir)
chess = cv2.cvtColor(chess, cv2.COLOR_BGR2RGB)
plt.imshow(chess)
```

```python
chess_copy = np.copy(chess)
chess_copy = cv2.cvtColor(chess_copy, cv2.COLOR_RGB2GRAY)
```

```python
#Harris corner detection
corners = cv2.cornerHarris(chess_copy, 2, 3, 0.04)
plt.imshow(corners, cmap = 'gray')
```

```python
corners_copy = np.copy(corners)
corners_copy = cv2.dilate(corners_copy, None)
plt.imshow(corners_copy)
```

```python
corners_copy.shape
```

```python
thresh = 0.01*corners_copy.max()
for i in range(corners_copy.shape[0]):
    for j in range(corners_copy.shape[1]):
        if corners_copy[j, i]>=thresh:
            cv2.circle(chess, (j, i), 1, (0, 255, 0), 1)
plt.imshow(chess)
```

```python
#Image contouring
#Process: grayscale image --> binary thresholding --> create a contour finder --> apply contour to image
hand_dir = 'C:/Users/rrsoo/AdvancedStuff/Udacity_CV/hand.jpg'
hand = cv2.imread(hand_dir)
hand = cv2.cvtColor(hand, cv2.COLOR_BGR2RGB)
plt.imshow(hand)
```

```python
hand_copy = np.copy(hand)
hand_copy = cv2.cvtColor(hand, cv2.COLOR_RGB2GRAY)
thresh = 245
retval, binary_hand = cv2.threshold(hand_copy, thresh, 255, cv2.THRESH_BINARY_INV)
#Objects that are being contoured must be white and background must be black
binary_hand_morph = np.copy(binary_hand)
binary_hand_morph = cv2.morphologyEx(binary_hand, cv2.MORPH_OPEN,(10,10))

f, (axis1, axis2) = plt.subplots(1, 2, figsize = (16, 8))

axis1.set_title("Original Binary Image")
axis1.imshow(binary_hand, cmap = 'gray')

axis2.set_title("Binary Image After Open Morphology")
axis2.imshow(binary_hand_morph, cmap = 'gray')

#Open morphology gets rid of stray noise
```

```python
contours, hierarchy = cv2.findContours(binary_hand_morph, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
final_copy = np.copy(hand)
countoured_image = cv2.drawContours(final_copy, contours, -1, (255, 0, 0), 2)
plt.imshow(countoured_image)
```

```python
#You can draw bounding boxes with contours!!!
x,y,w,h = cv2.boundingRect(contours[0])
cv2.rectangle(final_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(final_copy)
```

```python
#K-means segmentation
monarch_dir = 'C:/Users/rrsoo/AdvancedStuff/Udacity_CV/monarch.jpg'
monarch = cv2.imread(monarch_dir)
monarch = cv2.cvtColor(monarch, cv2.COLOR_BGR2RGB)
plt.imshow(monarch)
```

```python
#Excpects 2D array
monarch_copy = np.copy(monarch)
monarch_copy = np.float32(monarch.reshape(-1, 3))
```

```python
#Define stopping criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
k = 2
_, labels, centers = cv2.kmeans(monarch_copy, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_data = segmented_data.reshape(monarch.shape)
plt.imshow(segmented_data)
```
