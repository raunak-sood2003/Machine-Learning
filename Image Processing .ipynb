```python
from skimage import data
#Example images come from the submodule data
import cv2
#os works with the operating system, can use to get your own images
import matplotlib.pyplot as plt
#Used to show images
import numpy as np
astro = data.astronaut()
```

```python
#Original image
astro = data.astronaut()
plt.imshow(astro)
plt.show()
```

```python
#Flipping left/right, up/down
flippedud = np.flipud(astro)
flippedlr = np.fliplr(astro)
#plt.imshow(flippedud)
plt.imshow(flippedlr)
plt.imshow(flippedlr)
plt.show()
```

```python
#Blurring images using cv2.blur() and cv2.medianBlur()
average_blur = cv2.blur(astro, (5,5))
#1st method of blurring by averaging pixels
median_blur = cv2.medianBlur(astro,99)
#takes median instead of mean; can set blurriness by odd numbers; reduces noise
gaussian_blur = cv2.GaussianBlur(astro, (101,101),0)
plt.imshow(gaussian_blur)
plt.show()
```

```python
#Changing color-space using cv2.cvtColor()
gray_astro = cv2.cvtColor(astro, cv2.COLOR_RGB2GRAY)
hsv_astro = cv2.cvtColor(astro, cv2.COLOR_RGB2HSV)
plt.imshow(gray_astro, cmap = 'gray')
plt.show()
```

```python
np.asarray(gray_astro)
```

```python
#Object tracking e.g finding only the red stuff using cv2.inRange()
hsv_astro = cv2.cvtColor(astro, cv2.COLOR_RGB2HSV)
#U can look up the ranges online
lower_limit_red = np.array([10,100,20])
upper_limit_red = np.array([25,255,255])
red_astro = cv2.inRange(hsv_astro,lower_limit_red, upper_limit_red)
plt.imshow(red_astro)
plt.show()
```
